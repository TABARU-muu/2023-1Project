```python
import warnings
import pandas as pd
warnings.filterwarnings(action='ignore')

product_df=pd.DataFrame({
    "Client names":["정만수","조성훈","정만수","도민회","백성아","손정수","천미래","박찬수"],
    "Product names":["탕수육김치세트","탕수육김치세트","라면철면세트","된장푸딩세트","만두크림세트","타코청국장세트","짜장아이스크림세트","볶음세트"],
    "Purchase Values":[3,2,2,5,3,4,3,1],
})

product_df
```

```python
calc_df = product_df.groupby(by=["Product names"], as_index=False).sum()
calc_df
```

```python
calc_df["Purchase Values"] /= len(calc_df)
calc_df.sort_values(by = 'Purchase Values',ascending=False)
```
